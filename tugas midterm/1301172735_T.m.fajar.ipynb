#### Code

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#### trend analisys
```

```python
def trend(a):
    if a > -0.5 and a <= 0.5:
        return "Slight or No change"
    elif a > 0.5 and a <= 1:
        return "Slight Positive"
    elif a > -1 and a <= -0.5:
        return "Slight Negative"
    elif a > 1 and a <= 3:
        return "Positive"
    elif a > -3 and a <= -1:
        return "Negative"
    elif a > 3 and a <= 7:
        return "Among top gainers"
    elif a > -7 and a <= -3:
        return "Among top losers"
    elif a > 7:
        return "Bull run"
    elif a <= -7:
        return "Bear drop"
    
```

### Soal 4 (30 poin)
Lakukanlah analisis volatilitas dari ketiga harga saham tersebut selama tahun 2019, lalu tampilkan dan jelaskan hasil visualisasinya.

#### Code

```python
df_HangSeng2019 = pd.read_csv("Hang_Seng.csv") 
df_Nasdaq2019 = pd.read_csv("Nasdaq.csv") 
df_Nikkei2019 = pd.read_csv("Nikkei.csv") 

# merge Adj. close values
combined_df = pd.concat([df_HangSeng2019["Adj Close"], df_Nasdaq2019["Adj Close"], df_Nikkei2019["Adj Close"]], axis=1)
combined_df.columns = ["HangSeng", "Nasdaq", "Nikkei"]
```

```python
combined_df.dropna(inplace=True, axis = 0)
combined_df.head() 
```

```python
# store daily returns of all above stocks in a new dataframe 
pct_chg_df = combined_df.pct_change()*100
pct_chg_df.dropna(inplace = True, how = "any", axis = 0)

# plotting pairplot  
import seaborn as sns
sns.set(style = "ticks", font_scale = 1.25)
sns.pairplot(pct_chg_df)
```

```python
HangSeng_vol = pct_chg_df["HangSeng"].rolling(7).std()*np.sqrt(7)
HangSeng_vol.plot(figsize = (13, 6))
```

```python
Nasdaq_vol = pct_chg_df["Nasdaq"].rolling(7).std()*np.sqrt(7)
Nasdaq_vol.plot(figsize = (13, 6))
```

```python
Nikkei_vol = pct_chg_df["Nikkei"].rolling(7).std()*np.sqrt(7)
Nikkei_vol.plot(figsize = (13, 6))
```

- grafik nasdaq mengalami perubahan mencapai angka 20 keatas
